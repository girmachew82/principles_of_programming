- Set is collection of unique elements
- Elements are **unordered**
- It can be create by set(), not by {}

```python
vawels = {'a','e','i','o','u'}
consonants = {'b','c','d','f','g','h','j','k'}
print("Vowels union consonants =",vawels.union(consonants))
print("Vowels intersection consonants =",vawels.intersection(consonants))
print("Vowels difference consonants =",vawels.difference(consonants))
print("Vowels symmetric difference consonants =",vawels.symmetric_difference(consonants))
print("Vowels union consonants =",(vawels & consonants))
print("Vowels intersection consonants =",(vawels | consonants))
print("Vowels difference consonants =",(vawels - consonants))
print("Vowels symmetric difference consonants =",(vawels ^ consonants))
```

#**Summary**

**Lists** are used to store a sequence of values and are a mutable data structure, meaning their contents can be changed.
*   Lists are initialized using `list()` or `[]`.
- An empty list is considered a False value.
*   List elements are ordered and can be accessed by their index, either positive (starting from 0) or negative (starting from -1).
*   Accessing an index that is out of range will raise an `IndexError`.
*   Slicing is possible using a correct range, where the upper number of the range is excluded.
*   Lists can contain nested elements and elements of different data types.
*   Various built-in methods are available to manipulate lists. These include:
    *   `.append(x)` to add an element at the end of the list.
    *   `.insert(i, e)` to insert an element at a specific index.
    *   `.extend(iterable)` to add elements from an iterable to the end of the list.
    *   `.remove(element)` to remove the first occurrence of an element.
    *   `.pop([i])` to remove an element at a specific index (or the last element if no index is given).
    *   `.clear()` to remove all elements from the list.
    *   `del list[i]` to delete an element at a specific index.
    *   `.index(item)` to find the index of an element.
    *   `.count(item)` to count the occurrences of an element.
    *   `.sort()` to sort the list in ascending order, and `.sort(reverse=True)` for descending order.
    *   `.reverse()` to reverse the list.
    *   `.copy()` to create a copy of the list.
    *   `len(list)` to get the length of the list.
    *   `min()`, `max()`, and `sum()` for basic mathematical operations on numeric lists.
*   Lists can be manipulated using loops. `enumerate()` can access the index and value of list elements.
*   Lists can be unpacked into variables.
*   List arithmetic includes addition/concatenation (`+`) and multiplication/repetition (`*`).
*   Membership can be checked using the `in` and `not in` operators.
*   List comprehensions provide a compact way to create new lists by applying an operation to items in an existing sequence.
*   The `zip()` function can combine lists to create tuples.
*   List elements are mutable and can be changed.

**Tuples** are also used to store a sequence of values, but they are immutable, meaning their contents cannot be changed after creation.
*   Tuples are initialized using `tuple()` or `()`. An empty tuple is `()`.
*   If a tuple contains only one element, it must have a comma at the end.
*   Tuples have built-in methods for manipulation including:
    *   `.index(item)` to find the index of an element and `.count(item)` to count occurrences.
    *   `tuple(sorted(t))` for sorting, and `tuple(sorted(t, reverse=True))` for reverse sorting.
    *   Copying can be done through assignment.
    *   Concatenation (`+`) and repetition (`*`) are also supported.
*   Slicing and indexing are supported using `[]` and `[:]`.
*   The `len()` function returns the length of the tuple.
*   Membership can be checked using `in` and `not in`.
*   `min()`, `max()`, and `sum()` for basic mathematical operations on numeric tuples.
*   Tuples are immutable and do not support item assignment or the append operation.

**Dictionaries** store elements in key-value pairs, where keys must be unique.
*   Dictionaries are mutable and can be initialized using `{}` or `dict()`.
*   Dictionary elements are accessed through keys, and they maintain order from Python 3.6+.
*   Dictionaries have built-in methods and operations including:
    *   `.keys()` to get all the keys and `.values()` to get all the values.
    *   `.get(key)` or `dict[key]` to access a value by its key.
    *   `.items()` to get all key-value pairs.
    *   Sorting can be done using `dict(sorted(t))`.
    *   Reversing can be done using `dict(reversed(data.items()))`.
    *   Copying can be done through assignment or by using the `copy()` method.
*   Concatenation can be done using the `.update()` method or `**dict1, **dict2`.
*   Dictionary values can be repeated using a dictionary comprehension.
*   Slicing and indexing are done by first changing the dictionary to a list.
*    `len()` to get the length of the dictionary.
*   Membership can be checked using `in` and `not in` operators.
*   `min()`, `max()`, and `sum()` for basic mathematical operations on numeric dictionaries.

**Sets** are collections of unique, unordered elements.
*   Sets can be created using `set()`.
*   Sets are mutable.
*   Sets automatically remove duplicate data.
*   Set operations include union, intersection, difference, and symmetric difference.
*   Sets have built-in methods for manipulation including:
     *   `.add()` to add an element to the set and `.update()` to add multiple elements to a set.
    *   `.remove()`, `.clear()`, `.discard()`, and `del` are used for removing elements from the set.
    *   `.pop()` to get a random element from the set.
*   Sets are not subscriptable and cannot be accessed using an index.
*   `len()` gives the length of the set.
*   Membership is checked using `in` and `not in`.
*  `min()`, `max()`, and `sum()` for basic mathematical operations on numeric sets.
* Sets support mathematical set operations such as union (`|` or `.union()`), intersection (`&` or `.intersection()`), difference (`-` or `.difference()`), and symmetric difference (`^` or `.symmetric_difference()`).
